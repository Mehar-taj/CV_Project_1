## Importing Required Libraries

In this section, we import all libraries required for image processing and visualization.

```python
# OpenCV library for computer vision operations
import cv2

# NumPy for numerical and array operations
import numpy as np

# Matplotlib for displaying images
import matplotlib.pyplot as plt
```

## Loading the Image

The passport-size image is loaded from the project directory using OpenCV.

```python
# Read image from project folder
img = cv2.imread("passport.jpg")

# Verify image loading
if img is None:
    print("Image not found.")
else:
    print("Image loaded successfully.")
```

## Displaying the Original Image

OpenCV loads images in BGR format, while Matplotlib expects RGB format.

Therefore, we convert the image before displaying it.

```python
# Convert image from BGR to RGB
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display image
plt.figure(figsize=(6,6))

plt.imshow(image_rgb)

plt.title("Original Passport Image")

plt.axis("off")

plt.show()
```

## Image Transformation

Image transformation techniques help prepare images for analysis.

Transformations performed:

1. Grayscale Conversion
2. Image Resizing
3. Image Rotation

```python
# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Resize image
resized = cv2.resize(img, (300,300))

# Rotate image
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
```

```python
# Display transformed images

plt.figure(figsize=(12,4))

# Grayscale
plt.subplot(1,3,1) # plt.subplot(rows, columns, position) means the grayscale image display in the first position of a 1x3 grid
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

# Resized
plt.subplot(1,3,2) # the resize image display in the second position of a 1x3 grid
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized")
plt.axis("off")

# Rotated
plt.subplot(1,3,3) # the rotate image display in the third position of a 1x3 grid
plt.imshow(
    cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated")
plt.axis("off")

plt.show()
```

## Drawing Shapes and Adding Text

OpenCV provides tools for image annotation.

Annotations used:

- Rectangle
- Circle
- Text


```python
# Create image copy
img_copy = img.copy()

# Draw rectangle
cv2.rectangle(img_copy, (50,50), (200,200), (0,255,0), 3)
                        #straing point (x,y) ending point (x,y) color (B,G,R) thickness
# Draw circle
cv2.circle(img_copy, (300,300), 50, (0,0,255), 3)
                    #center point, radius, color, thickness
# Add text
cv2.putText(img_copy, "OpenCV Demo", (30,300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
                    # Demo text, starting point of the text, font size, font size, color,thickness

# Display annotated image

plt.figure(figsize=(6,6))

plt.imshow(cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB))

plt.title("Annotated Image")

plt.axis("off")

plt.show()
```

## Image Filtering

Filtering improves image quality and extracts useful features.

Techniques used:

- Gaussian Blur
- Canny Edge Detection

```python
# Apply Gaussian Blur
blurred = cv2.GaussianBlur(img,(15,15),0)

# Convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect edges
edges = cv2.Canny(gray_img,100, 200)

# Display filtering results

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
plt.title("Gaussian Blur")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edges, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")

plt.show()
```

## Face Detection using Haar Cascade Classifier

Face detection is performed using OpenCV's pre-trained Haar Cascade model.

The classifier scans the image and identifies facial regions.

```python
# Load Haar Cascade model
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades +
    "haarcascade_frontalface_default.xml"
)

# Convert image to grayscale
gray_face = cv2.cvtColor(
    img,
    cv2.COLOR_BGR2GRAY
)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray_face,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30,30))

print("Number of Faces Detected:",len(faces))
```

```python
# Create image copy
img_faces = img.copy()

# Draw rectangles around faces
for (x,y,w,h) in faces:

    cv2.rectangle(
        img_faces,
        (x,y),
        (x+w,y+h),
        (255,0,0),
        2
    )

# Display face detection result

plt.figure(figsize=(6,6))

plt.imshow(
    cv2.cvtColor(
        img_faces,
        cv2.COLOR_BGR2RGB
    )
)

plt.title("Detected Faces")

plt.axis("off")

plt.show()
```
